```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
COLS= ["Area", "Perimeter", "Compactness", "Length", "Wdith", "Asymmetry", "Groove", "Class"]
df = pd.read_csv('/content/seeds_dataset.txt', sep='\s+', header=None, names=COLS)
display(df.head())
df
```

```python
for i in range(len(COLS)-1):
  for j in range (i+1 , len(COLS)-1):
    x_label = COLS[i]
    y_label = COLS[j]
    sns.scatterplot(x=x_label, y=y_label, data=df, hue= "Class")
    plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
x ="Perimeter"
y ="Asymmetry"
X = df[[x,y]].values
```

```python
Kmeans = KMeans(n_clusters=3)
Kmeans.fit(X)
```

```python
clusters = Kmeans.labels_
```

```python
clusters
```

```python
df['Class'].values
```

```python
cluster_df =  pd.DataFrame(np.hstack((X,clusters.reshape(-1, 1))), columns=[x ,y ,"class"])
```

```python
#ORIGINAL_CLASSES
sns.scatterplot(x=x , y=y , hue = 'Class',data = df)
plt.plot()
```

```python
#K-MEANS CLASSES
sns.scatterplot(x=x , y=y , hue = 'class',data = cluster_df)
plt.plot()
```

#HIGHER DIMENSIONS

```python
X= df[COLS[:-1]].values
```

```python
kmeans = KMeans(n_clusters = 3).fit(X)
cluster_df = pd.DataFrame(np.hstack((X, kmeans.labels_.reshape(-1 ,1))), columns=df.columns)
```

```python
# K Means classes
sns.scatterplot(x=x, y=y, hue='Class', data=cluster_df)
plt.plot()
```

```python
# Orignial  classes
sns.scatterplot(x=x, y=y, hue='Class', data=df)
plt.plot()
```
